```python
channels = 14
# Initialize an empty list to store the FFT magnitudes for each chunk
fft_rows = []

# Iterate over each chunk of 128 rows
for i in range(1, 2):        
    fft_magnitudes = []
    
    # Iterate over each column
    for col in range(0, channels):
        # Extract the chunk of 128 rows for the current column
        signal = df.iloc[sample_rate * (i - 1):sample_rate * i, col].copy(deep=True)    # window size of 128 "time samples"  from 1 channels
        print(signal)
        signal -= signal.mean()
        print(signal)

        # Compute the FFT of the extracted rows
        fft_output = np.fft.fft(signal)
        print(fft_output)
        
        # Compute the magnitude of the FFT result
        fft_magnitude = np.abs(fft_output)
        print(fft_magnitude)
        print(fft_magnitudes)

        
        # Append the magnitudes to the list for this chunk
        fft_magnitudes.extend(fft_magnitude)
        print(fft_magnitudes)
    # Append the FFT magnitudes of this chunk as a new row in the fft_rows list
    fft_rows.append(fft_magnitudes)
    
# Convert the list of FFT rows to a DataFrame
new_table = pd.DataFrame(fft_rows)
# should be 1792 coulmns (128 samples* 14 channels) , rows = 116 (sec) 
new_table
```

```python
channels = 14
# Initialize an empty list to store the FFT magnitudes for each chunk
fft_rows = []

# Iterate over each chunk of 128 rows
for i in range(1, total_duration):        
    fft_magnitudes = []
    
    # Iterate over each column
    for col in range(0, channels):
        # Extract the chunk of 128 rows for the current column
        signal = df.iloc[sample_rate * (i - 1):sample_rate * i, col]    # window size of 128 "time samples"  from 1 channels
        print(col)
        print(signal)
        average_signal = signal
        average = average_signal.mean()
        average_signal -= average
        print(average_signal)

        # Compute the FFT of the extracted rows
        fft_output = np.fft.fft(average_signal)
        
        # Compute the magnitude of the FFT result
        fft_magnitude = np.abs(fft_output)
        
        # Append the magnitudes to the list for this chunk
        fft_magnitudes.extend(fft_magnitude)
    
    # Append the FFT magnitudes of this chunk as a new row in the fft_rows list
    fft_rows.append(fft_magnitudes)

# Convert the list of FFT rows to a DataFrame
new_table = pd.DataFrame(fft_rows)
# should be 1792 coulmns (128 samples* 14 channels) , rows = 116 (sec) 
new_table
```

```python
#building the eye state of 128 combined samples
#The 14 column indicate that the eye close
#If sum == 0  --> class of batch is "0" (eye-open)
#If sum == 128  --> class of batch is "1" (eye-closed)
#if sum < 128  --> class of batch is "2" (blinking eye)

list = []
for i in range(1,total_duration):
    sum_of_batch =df.iloc[sample_rate * (i - 1):sample_rate * i, 14].sum()  # sum of window size of 128 "time samples"  of classes
    list.append(sum_of_batch)
sum = pd.DataFrame(list)
sum_data = {'value' :list}
sum_data_df = pd.DataFrame(sum_data)

new_rows = []
for index, row in sum_data_df.iterrows():
    value = row['value']
    if value == 0:
        new_row = [0, 1, 0]
    elif value == 128:
        new_row = [1, 0, 0]
    else:
        new_row = [0, 0, 1]
    new_rows.append(new_row)
class_df = pd.DataFrame(new_rows)
print(class_df)
```

```python
#expand new_table with 3 additional columns for classes
for col in range(3):
    new_table[col + 1792] = np.nan  # Start adding columns from column index 128

# Iterating over class_df
for i in range(116):
    new_table.iloc[i, -3:] = class_df.iloc[i]
new_table
```
